```python
from google.colab import files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

```

```python
uploaded = files.upload()
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
```

```python
x_train = df["year"].values.reshape(-1,1)
y_train = df["per capita income (US$)"].values

```

```python
plt.scatter(x_train,y_train,color="red",marker="x")
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.title("canada_per_capita_income")
plt.show()
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
print("Slope (coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
year = 2025
predicted_income = model.predict([[year]])  # input must be 2D
print(f"Predicted income for {year}: ${predicted_income[0]:.2f}")
```

```python
plt.scatter(x_train, y_train, color='blue', label='Actual data')
plt.plot(x_train, model.predict(x_train), color='red', label='Regression line')
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Linear Regression Fit")
plt.legend()
plt.grid(True)
plt.show()
```
